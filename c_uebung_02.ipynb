```c++
// Modified to 8 bit
// https://www.geeksforgeeks.org/binary-representation-of-a-given-number/
void bin(short n)
{
    short i;
    printf("0b");
    for (i = 1 << 7; i > 0; i = i / 2)
        (n & i) ? printf("1") : printf("0");
    printf("\n");
}
```

```c++
short val = (5 | 12);
printf("%d\n", val);
printf("%x\n", val);
bin(val);
```

```c++
short val = (2 & 4);
printf("%d\n", val);
printf("%x\n", val);
bin(val);
```

```c++
val = ( 2 & (5 > 3) );
printf("%d\n", val);
printf("%x\n", val);
bin(val);
```
