```c++
a = 1; b = 10; c = 0;
for (c = 0; c <= 5; c++) {
    if(a < 3){
        b = b - 1;
    }
    a = a + 1;
    b = b + 2;
}
printf("a: %d\tb: %d\tc: %d", a, b, c);
```

```c++
char letter = 'B';  x = 0;
switch(letter){
    case 'A': x= 1; break;
    case 'B': x= 1;
    case 'C': x= 3; break;
    default: x = 4;
} 
printf("x: %d", x);
```

```c++
int f(int a, int b){
    a = a * b;
    b = 0;
    return a;
}
```

```c++
a = 4; x = 3; int z = f(a, x);
printf("a: %d\tx: %d\tz: %d", a, x, z);
```

```c++
short g(int *p, short z){
    (*p)--;
    return z + 1;
}
```

```c++
int c = 0; short y= 32767;
short n = g(&c, y);
printf("c: %d\ty: %d\tn: %d", c, y, n);
```
